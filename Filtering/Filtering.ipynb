```python
import cv2 as cv
from google.colab.patches import cv_imshow
import numpy as np

img = cv.imread('/content/gdrive/MyDrive/MachineVision/Re2/DOG.jpg')
blur3= cv.blur(img,(3,3))
blur5= cv.blur(img,(5,5))
blur7= cv.blur(img,(7,7))
cv_imshow(img)
cv_imshow(blur3)
cv_imshow(blur5)
cv_imshow(blur7)

cv.imwrite('/content/gdrive/MyDrive/MachineVision/Re2/blur3.jpg',blur3)
cv.imwrite('/content/gdrive/MyDrive/MachineVision/Re2/blur5.jpg',blur5)
cv.imwrite('/content/gdrive/MyDrive/MachineVision/Re2/blur7.jpg',blur7)
```

```python
blur = cv.blur(img, ksize=(7,7))
detail = np.int32(img) - np.int32(blur)
shapened_img2 = np.int32(img) + 2*detail
shapened_img5 = np.int32(img) + 5*detail
shapened_img10 = np.int32(img) + 10*detail
cv_imshow(img)

cv_imshow(shapened_img2)
cv_imshow(shapened_img5) 
cv_imshow(shapened_img10) 
#cv.imwrite('/content/gdrive/MyDrive/MachineVision/Re2/shapened_7img2.jpg',shapened_img2)
#cv.imwrite('/content/gdrive/MyDrive/MachineVision/Re2/shapened_7img5.jpg',shapened_img5)
cv.imwrite('/content/gdrive/MyDrive/MachineVision/Re2/shapened_7img10.jpg',shapened_img10)
```

```python
kernel1d	=	cv.getGaussianKernel(7, 2)
kernel2d = np.outer(kernel1d, kernel1d.transpose())

filterimg	=	cv.filter2D(img, -1, kernel2d)
cv_imshow(filterimg)
filterimg2	=	cv.filter2D(img, -1, kernel1d)
filterimg3	=	cv.filter2D(filterimg2, -1, kernel1d.transpose())
cv_imshow(filterimg3)
cv.imwrite('/content/gdrive/MyDrive/MachineVision/Re2/filterimg3.jpg',filterimg3)
```

```python
import cv2 as cv
import numpy as np
from google.colab.patches import cv_imshow

def addsaltandpeppernoise(image,ratio):
  row,col,ch = image.shape
  out = np.copy(image)
  # Salt mode
  num_salt = np.ceil(image.size * ratio)
  coords = [np.random.randint(0, i - 1, int(num_salt)) for i in image.shape]
  out[coords] = 255
  # Pepper mode
  num_pepper = np.ceil(image.size * ratio)
  coords = [np.random.randint(0, i - 1, int(num_pepper)) for i in image.shape]
  out[coords] = 0
  return out
```

```python
saltimg002 = addsaltandpeppernoise(img, ratio=0.02)
saltimg01 = addsaltandpeppernoise(img, ratio=0.1)
saltimg25 = addsaltandpeppernoise(img, ratio=0.25)

median_002img = cv.medianBlur(saltimg002, ksize=7)
median_01img = cv.medianBlur(saltimg01, ksize=7)
median_25img = cv.medianBlur(saltimg25, ksize=9)
cv_imshow(saltimg002)
cv_imshow(saltimg01)
cv_imshow(saltimg25)
#cv.imwrite('/content/gdrive/MyDrive/MachineVision/Re2/saltimg002.jpg',saltimg002)
#cv.imwrite('/content/gdrive/MyDrive/MachineVision/Re2/saltimg01.jpg',saltimg01)
#cv.imwrite('/content/gdrive/MyDrive/MachineVision/Re2/saltimg25.jpg',saltimg25)


cv_imshow(median_002img)
cv_imshow(median_01img)
cv_imshow(median_25img)

cv.imwrite('/content/gdrive/MyDrive/MachineVision/Re2/7median_002img.jpg',median_002img)
cv.imwrite('/content/gdrive/MyDrive/MachineVision/Re2/7median_01img.jpg',median_01img)
cv.imwrite('/content/gdrive/MyDrive/MachineVision/Re2/9median_25img.jpg',median_25img)
```
